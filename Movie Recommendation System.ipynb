```python
import ast # it is used to convert string in list
```

```python
def change(sent):
    a=[]
    for i in ast.literal_eval(sent):
        a.append(i['name'])
    return a
```

```python
new_movies['genres']=new_movies['genres'].apply(change)
```

```python
# samre function used for keyword
new_movies['keywords']=new_movies['keywords'].apply(change)
```

```python
# For cast
# we have to create new function because from this we want only first three dictionary
def change1(sent):
    a=[]
    counter=0
    for i in ast.literal_eval(sent):
        if counter!=3:
            a.append(i['name'])
            counter+=1
        else:
            break
    return a
```

```python
new_movies['cast']=new_movies['cast'].apply(change1)
```

```python
def get_director(sent):
    a=[]
    
    for i in ast.literal_eval(sent):
        if i['job']=='Director':
            a.append(i['name'])
            break
        
            
    return a
```

# Count Vectorization

```python
from sklearn.feature_extraction.text import CountVectorizer
cv=CountVectorizer(max_features=5000,stop_words='english')
```

```python
vectors=cv.fit_transform(movies['about']).toarray()
```

I have two issues in this like 
1. For ex actions or action have same meanings but have diff words
2. Digits

So I Am Applying Stemming Here

```python
import nltk
```

```python
from nltk.stem.porter import PorterStemmer
ps=PorterStemmer()
```

```python
def stem(word):
    y=[]
    for i in word.split():
        y.append(ps.stem(i))
        string= ' '.join(y) 
    return string
```

```python
movies['about']=movies['about'].apply(stem)
```

```python
vectors=cv.fit_transform(movies['about']).toarray()
```

Now we have to calculate the distance of every movie with another movie , so we calculate Cosine Similarity, means calculate the angle between them. 

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
similarity_matrix=cosine_similarity(vectors)
```

```python
cosine_similarity(vectors).shape
```

```python
similarity_matrix[0]
```

```python
# Now we have to make a function of recommendation
def recommend(movie):
    movie_index=movies[movies['title']==movie].index[0]
    similarity=similarity_matrix[movie_index]
    recommend_movie=sorted(list(enumerate(similarity)),reverse=True,key=lambda x:x[1])[1:6]
    
    for i in recommend_movie:
        print(movies.iloc[i[0]].title)
        
    
    
```

```python
recommend('Avatar')
```

```python
recommend('Independence Day')
```
